```python
# ------------------------------------------------------------
# 1) จัดการข้อมูลพื้นฐาน
# ------------------------------------------------------------
import pandas as pd          # ใช้อ่านไฟล์ CSV / จัดการ DataFrame
import numpy as np           # ใช้คำนวณตัวเลข, จัดการ array, แทน missing

# ------------------------------------------------------------
# 2) แบ่งข้อมูล Train / Test
# ------------------------------------------------------------
from sklearn.model_selection import train_test_split
# ใช้แบ่งข้อมูลเป็นชุดฝึกสอน (train) และทดสอบ (test)

# ------------------------------------------------------------
# 3) การทำ Preprocessing (เตรียมข้อมูล)
# ------------------------------------------------------------
from sklearn.preprocessing import StandardScaler, OneHotEncoder
# StandardScaler = ทำให้ตัวเลขอยู่ในสเกลเดียวกัน
# OneHotEncoder = แปลงข้อมูลหมวดหมู่เป็นตัวเลขแบบ one-hot

from sklearn.compose import ColumnTransformer
# ใช้กำหนดว่าคอลัมน์ไหนใช้ scaler, คอลัมน์ไหนใช้ one-hot

from sklearn.impute import SimpleImputer
# ใช้เติมค่า missing เช่น ค่าเฉลี่ย, ค่ามัธยฐาน, ค่าที่พบบ่อย

# ------------------------------------------------------------
# 4) Pipeline (รวมขั้นตอน preprocessing + model)
# ------------------------------------------------------------
from sklearn.pipeline import Pipeline
# ใช้ต่อขั้นตอนเป็นสายเดียว เช่น (เติม missing → scale → model)

# ------------------------------------------------------------
# 5) ตัวโมเดล
# ------------------------------------------------------------
from sklearn.linear_model import LinearRegression
# ใช้สร้างโมเดล Linear Regression (เส้นตรง)

# ------------------------------------------------------------
# 6) การประเมินผลโมเดล
# ------------------------------------------------------------
from sklearn.metrics import mean_squared_error
# ใช้วัดความผิดพลาดของโมเดลแบบ MSE

# 1) โหลดข้อมูล
df = pd.read_csv('housing (1).csv')

# 2) แยก Feature / Target
X = df.drop('median_house_value', axis=1)
y = df['median_house_value']

# 3) แยกประเภทคอลัมน์
num_col = X.select_dtypes(include=['int64', 'float64']).columns
cat_col = X.select_dtypes(include=['object']).columns

# 4) Preprocessing: Impute + Scale + OneHot
preprocess = ColumnTransformer(
    transformers=[
        ('num',
         Pipeline([
             ('imputer', SimpleImputer(strategy='median')),
             ('scaler', StandardScaler())
         ]),
         num_col),

        ('cat',
         Pipeline([
             ('imputer', SimpleImputer(strategy='most_frequent')),
             ('ohe', OneHotEncoder(sparse_output=False, handle_unknown='ignore'))
         ]),
         cat_col)
    ]
)

# 5) แปลงข้อมูล
X_processed = preprocess.fit_transform(X)

# 6) Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y, test_size=0.2, random_state=42
)

# 7) โมเดล
model = LinearRegression()
model.fit(X_train, y_train)

# 8) ทำนาย
y_pred = model.predict(X_test)

# 9) ประเมินผล
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("MSE:", mse)
print("RMSE:", rmse)
```
